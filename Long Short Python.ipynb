```python
#Libraries
import pandas as pd
import numpy as np

#import data
df1 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLV_Health_Care")
df2 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLP_Consumer_Staples")
df3 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLK_Information_Technology")
df4 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLI_Industrials")
df5 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLF_Financials")
df6 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLE_Energy")
df7 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLB_Materials")
df8 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLY_Consumer_Discretionary")
df9 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "XLU_Utilities")
df10 = pd.read_excel("/Users/gracezhang/Documents/S&P500 Long-Short Strategy.xlsx", sheet_name = "SPY_S&P500")

#stitch the sectors close data together into one DF
data = pd.DataFrame()
data["Date"] = df1.iloc[:,0]
data["Healthcare"] = df1.iloc[:,1]
data["Consumer Staples"] = df2.iloc[:,1]
data["Information Technology"] = df3.iloc[:,1]
data["Industrials"] = df4.iloc[:,1]
data["Financials"] = df5.iloc[:,1]
data["Energy"] = df6.iloc[:,1]
data["Materials"] = df7.iloc[:,1]
data["Consumer Discretionary"] = df8.iloc[:,1]
data["Utilities"] = df9.iloc[:,1]
data["S&P500"] = df10.iloc[:,1]

data.head()
```

```python

#processing the Data to get Sept returns
#set Date column that was text to a Date data type
data['Date'] = pd.to_datetime(data['Date'])

#get the month of a given date
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

#get grouping column
data['Year_Month_Symbol_Group'] = (data['Year'] * 100 + data['Month']).astype(str)

#get first and last days within a year-month group and put them into a max and min dataframe
data_max = data[
    data['Date'] == data.groupby('Year_Month_Symbol_Group')['Date'].transform('max')
    ]
data_max['Month End?'] = "Yes"

data_min = data[
    data['Date'] == data.groupby('Year_Month_Symbol_Group')['Date'].transform('min')
    ]
data_min['Month End?'] = "No"

data.head()
```

```python
#combine the max and min date dataframes together so that we only take max and min d
data_combined = pd.concat([data_max, data_min], axis=0)
data_combined.head()

#set the original dataframe to this combined one
data = data_combined
data = data.sort_values(by=['Date'])
data.head(1)
```

```python


data["Monthly Return Healthcare"] = np.nan
data["Monthly Return Consumer Staples"] = np.nan
data["Monthly Return Information Technology"] = np.nan
data["Monthly Return Industrials"] = np.nan
data["Monthly Return Financials"] = np.nan
data["Monthly Return Energy"] = np.nan
data["Monthly Return Materials"] = np.nan
data["Monthly Return Consumer Discretionary"] = np.nan
data["Monthly Return Utilities"] = np.nan
data["Monthly Return S&P500"] = np.nan

data.head()




```

```python
#calculate returns based on month end or not
data_len = len(data)

def returns (startp, endp):
    return (endp / startp) - 1

#indexing setup
month_end_index = 14

#apply the returns formula to all returns columns
for x in range(15, 25):
    for i in range(data_len):
        if (data.iloc[i, month_end_index] == "Yes"): #if month end?
            data.iloc[i, x] = returns(data.iloc[i-1, x-month_end_index] #calculate returns on month start p
                                    , data.iloc[i, x-month_end_index]) #and month end p

data = data[data.iloc[:,month_end_index] == "Yes"]
data.head()
```

```python
selection = list(range(15,25))
selection.append(0)
selection.sort()
print(selection)
```

```python
returns = data.iloc[:, selection]
returns.head(26)
```

```python
#get a df copy of dimensions with the returns df
avg_df = returns.copy()
avg_df.columns = [
    "Date",
    "Healthcare (%)",
    "Consumer Staples (%)",
    "Information Technology (%)",
    "Industrials (%)",
    "Financials (%)",
    "Energy (%)",
    "Materials (%)",
    "Consumer Discretionary (%)",
    "Utilities (%)",
    "S&P500 (%)"]
    

#parse the df
avg_len_index = len(avg_df)

for n in range(1, 11):
    for m in range(5, avg_len_index): 
        #print(m, n)
        yr1 = returns.iloc[m-5, n] + 1
        yr2 = returns.iloc[m-4, n] + 1
        yr3 = returns.iloc[m-3, n] + 1
        yr4 = returns.iloc[m-2, n] + 1
        yr5 = returns.iloc[m-1, n] + 1
        avg_df.iloc[m, n] = ((yr1 + yr2 + yr3 + yr4 + yr5)/5 - 1) * 100 #cagr instead pow(yr1 * yr2 * yr3 * yr4 * yr5, 0.2)

#drop the first 4 years that don't get processed
avg_df = avg_df.iloc[5:]

avg_df.head(21)
```

```python
avg_df_wo_date = avg_df.copy().iloc[:, 1:10]
row_index = len(avg_df_wo_date)

trading_returns = avg_df.copy().iloc[:, 0:2]
trading_returns.columns = ["Date", "Strategy Returns (%)"]
trading_returns["Strategy Returns (%)"] = 0

for i in range(0, row_index):

    #get a list for each row of returns in the non-date avg df
    row = avg_df_wo_date.iloc[i, 0:10].values.flatten().tolist()

    #create a tuple of indices
    indices = range(1, 10)

    #sort the tuples based on the avg_return and sort to get first and last 2
    sorted_list = sorted(tuple(zip(row, indices)))

    #indices to return to returns df
    i1 = sorted_list[0][1]
    i2 = sorted_list[1][1]
    
    print(i1, i2)

    #insert data into the trading strategy df
    y = returns.iloc[i+5, 10] - returns.iloc[i+5, i2] - returns.iloc[i+5, i1] #long and shorts
    print(y)
    trading_returns.iloc[i, 1] = y/3 * 100

trading_returns["S&P500 (%)"] = returns.iloc[5:,10] * 100
trading_returns.head(21)
#avg_df_wo_date.head()
```

```python
#Calculate Cumulative Returns for Trading Strategy and S&P500

trading_returns['Cumulative Strategy Returns (%)']= ((1 + trading_returns['Strategy Returns (%)']/100).cumprod() - 1) * 100
trading_returns['Cumulative S&P500 Returns (%)']= ((1 + trading_returns['S&P500 (%)']/100).cumprod() - 1) * 100
trading_returns.head(21)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Plot Strategy cumulative returns
plt.plot(trading_returns['Date'], trading_returns['Cumulative Strategy Returns (%)'], 
         marker='o', label='Cumulative Strategy Returns (%)')

# Plot S&P500 cumulative returns
plt.plot(trading_returns['Date'], trading_returns['Cumulative S&P500 Returns (%)'], 
         marker='s', label='Cumulative S&P500 Returns (%)')

# Labels and title
plt.xlabel('Date')
plt.ylabel('Cumulative Strategy Returns (%)')
plt.title('Cumulative Strategy vs S&P500 Trading Returns')
plt.xticks(trading_returns['Date'])  # ensure all years show on x-axis
plt.xticks(rotation=90) # Rotate labels by 90 degrees (vertical)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```
